# Indicators

```python
# Returns a DataFrame with the maximum value for each column and the reference
# to the country that has that value
# if drop_str is True, it drops every column that has non-numeric values
def Find_Max(df_data, col_index, favorable, drop_str=True):
    
    #Creating a new dataframe
    df_max = pd.DataFrame(columns=df_data.columns)

    
    for col in df_data:
        if type(max(df_data.loc[:,col])) == int or type(max(df_data.loc[:,col])) == float:
            df_max.loc["Max_Value",col] = max(df_data.loc[:,col])
        else:
            if drop_str == True:
                df_max.drop(col,axis=1,inplace=True)
            continue
        df_sorted = df_data.sort_values(by=col,ascending=False).reset_index()
        df_max.loc["Max_Object",col] = df_sorted.loc[0,col_index]
        
        df_max.loc["Favorable",col] = favorable[col]
    
    #Returns a dataframe with the max value per each column and what row has it
    return df_max


# Return a DataFrame with scores compared to the best score instead of values
def score_data(df_data, df_max):
    
    for col in df_data:
        for index, row in df_data.iterrows():

                if not pd.isnull(df_data.loc[index, col]):
                    df_data.loc[index,col] = (df_data.loc[index, col] / df_max.loc["Max_Value",col]) * 100
                    if df_max.loc["Favorable", col] == "Ascending":
                        df_data.loc[index ,col] = abs(df_data.loc[index ,col] - 100)

    return df_data

def create_score_column(df_data, dict_cols, col_name):
    
    if sum(dict_cols.values()) != 1:
        return "Error, the sum of the weights must be equal to 1"
    for index, row in df_data.iterrows():
        score_indicator = 0
        for col in dict_cols.keys():
            score_indicator += df_data.loc[index, col] * dict_cols[col]
        
        df_data.loc[index, col_name] = score_indicator
    return df_data
```
